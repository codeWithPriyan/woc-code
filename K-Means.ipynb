```python
#Importing the csv files into dataframes (both train and test)

dftrain = pd.read_csv("/content/drive/MyDrive/WOC/emnist-letters-train.csv", header=None)
dftest = pd.read_csv("/content/drive/MyDrive/WOC/emnist-letters-test.csv", header=None)
```

```python
#Functions

def importing(dataframe, examples):
    X = dataframe.iloc[:examples,1:]
    features = X.shape[1]
    examples = X.shape[0]
    Y = dataframe.iloc[:examples,0].to_numpy().reshape(examples, 1)
    return X, Y, features, examples

Xtrain, Ytrain, trainfeatures, trainexamples = importing(dftrain, 88800)
Xtest, Ytest, testfeatures, testexamples = importing(dftest, 14800)

def scaling(X):
    X = X.div(255)
    X = X.to_numpy()
    return X

Xtrain = scaling(Xtrain)
Xtest = scaling(Xtest)
```

```python
def kmeans(k, iteration):
  centres = np.random.randn(k,testfeatures)
  for j in range(iteration):
    cluster = {}
    for m in range(k):
      cluster[m] = []
    for i in range(testexamples):
      X = np.repeat(Xtest[i,:].reshape(1,testfeatures), k, axis=0)
      diff = X - centres
      dist = np.sqrt(np.sum((diff)**2, axis=1))
      minarg = np.argmin(dist)
      cluster[minarg].append(i)
    centresupdate = np.empty((k, testfeatures))
    for n in range(k):
      centresupdatetemp = np.empty((len(cluster[n]), testfeatures))
      for o in range(len(cluster[n])):
        centresupdatetemp[o,:] = np.array(Xtest[cluster[n][o]])
      if len(cluster[n]) != 0:
        centresupdate[n] = np.sum(centresupdatetemp, axis=0)/len(cluster[n])
    centres = centresupdate
  return centres, cluster
```

```python
centres, cluster = kmeans(19,100)
```
