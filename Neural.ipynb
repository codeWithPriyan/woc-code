```python
class NeuralNetwork:

  def __init__(self, nohdn, layersize, X):
    self.nohdn = nohdn
    self.layersize = layersize
    self.X = X
    self.inputftrs = self.X.shape[0]
    self.examples = self.X.shape[1]
    self.layers = [self.inputftrs] + self.layersize + [26]

  def initiation(self):
    self.weights = []
    self.bias = []
    for i in range(self.nohdn + 1):
      weight = np.random.randn(self.layers[i+1], self.layers[i])
      self.weights.append(weight)
      bias = np.random.randn(self.layers[i+1], 1)
      self.bias.append(bias)
    return self.weights, self.bias

  def onevsall(self, Y):
    Yzeros = np.zeros((self.examples, 26))
    for i in range(self.examples):
      val = Y[i] - 1
      Yzeros[i][val] = 1
    Y = Yzeros
    return Y.T
  
  def sigmoid(self, X, theta, bias):
    z = np.dot(theta, X) + bias
    return 1/(1+np.exp(-z))

  def activations(self, X):
    A = [X]
    for i in range(self.nohdn + 1):
      Xhidden = self.sigmoid(A[i], self.weights[i], self.bias[i])
      A.append(Xhidden)
    A.reverse()
    return A

  def model(self, lr, iteration):
    cost_list = []
    self.weights, self.bias = self.initiation()
    Y = self.onevsall(Ytrain_emnist)
    for i in range(iteration):
      activations = self.activations(self.X)
      cost = (-1/self.examples) * np.sum(Y*np.log(activations[0]) + (1-Y)*np.log(1-activations[0]))
      print(i+1, end=' ')
      print(cost)
      cost_list.append(cost) 
      delta_list = [0 for i in range(self.nohdn + 1)]
      der_weight_list = [0 for i in range(self.nohdn + 1)]
      der_bias_list = [0 for i in range(self.nohdn + 1)]
      delta_list[0] = activations[0] - Y
      der_weight_list[0] = 1/self.examples * np.dot(delta_list[0], activations[1].T)
      der_bias_list[0] = 1/self.examples * np.sum(delta_list[0], axis=1, keepdims=True)
      for j in range(self.nohdn):
        delta_list[j+1] = 1/self.examples * np.dot(self.weights[-1-j].T, delta_list[j]) * activations[1+j] * (1 - activations[1+j])
        der_weight_list[j+1] = 1/self.examples * np.dot(delta_list[1+j], activations[2+j].T)
        der_bias_list[j+1] = 1/self.examples * np.sum(delta_list[1+j], axis=1, keepdims=True)
      der_weight_list.reverse()
      der_bias_list.reverse()
      for k in range(self.nohdn + 1):
        self.weights[k] -= lr * der_weight_list[k]
        self.bias[k] -= lr * der_bias_list[k]
    return activations[0], self.weights, self.bias

  def accuracy_final(self, X, Y):
    activations_test = self.activations(X)
    Y_pred = np.argmax(activations_test[0], axis=0) + 1
    Y_check = (Y_pred - Y)
    accuracy = Y_check.tolist().count(0)/len(Y)*100
    return accuracy, Y_check, Y_pred
```
